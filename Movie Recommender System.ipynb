```python
import pandas as pd 
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from bs4 import BeautifulSoup
from requests import get
movie = []
director = []
genres = []
cast = []
ratings = []
overview = []
```

```python
sample = 1
i=1
for j in range(0,200):
    url = "https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&sort=user_rating,desc&start="+str(sample)+"&ref_=adv_nxt"  # sample is a variable
    #chrome = webdriver.Chrome()
    response = get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    movie_containers = soup.find_all('div',class_='lister-item mode-advanced')
    for k in range(0,len(movie_containers)):
        movie_name = movie_containers[k].find('div',class_='lister-item-content')
        movie.append(movie_name.h3.a.text)
        ratings_bar = movie_containers[k].find('div',class_='ratings-bar')
        ratings.append(ratings_bar.div.strong.text)
        genre_container = movie_name.find('p',class_='text-muted')
        genre_class = movie_name.find('span',class_='genre')
        genres.append(genre_class.text)
        #overview.append(overview_text.text)
        directors = movie_name.find('p',class_="")
        director.append(directors.a.text)
        coverview = movie_name.select('p')[1].get_text(strip=True)
        overview.append(coverview)
    print(i)
    i+=1
    #chrome.close()
    sample+=50
```

```python
data = {
    "Title": movie,
    "Ratings": ratings,
    "Genre": genres,
    "Director": director,
    "Overview": overview
}
```

```python
df = pd.DataFrame(data)
df['Ratings'] = df['Ratings'].astype(str)
```

```python
df['Director'] = df['Director'].str.replace(" ", "")
df
```

```python
df['Genre'] = df['Genre'].str.replace("\n", "")
df
```

```python
df['Tags'] = df['Genre'] + df['Director']+ ' ' + df['Overview'] + ' ' + df['Ratings']
```

```python
df = df[['Title','Tags']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 9999,stop_words='english')
```

```python
df['Tags'] = df['Tags'].apply(lambda x:x.lower())
#df['Tags'] = df['Tags'].apply(lambda x:x.split())
```

```python
df
```

```python
vectors = cv.fit_transform(df['Tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
df['ID'] = df.index+1
```

```python
df
```

```python
sorted(enumerate(similarity[0]),reverse = True,key=lambda x:x[1])[1:6]
```

```python
len(cv.get_feature_names_out())
```

```python
def recommend(movie):
    movie_index = df[df['Title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted((enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].Title)
```

```python
recommend('Batman')
```
